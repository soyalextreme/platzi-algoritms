```
import random


global debbug
debbug = True

def print_years(l):
    for row in l:
        print(row)

def print_ages(l):
    ages = []
    for i in range(len(l[0])):
        ages.append(l[1][i] - l[0][i])
    print(ages)


def merge(l):
    years = []
    for i in range(2):
        for j in range(len(l[0])):
            if i == 0:
                years.append(l[i][j])
            else:
                years.append(l[i][j]+ 1.5)
                
    years.sort()
    return years


def year_most_people_alive(l, n_people=0, h_pyear=[0,0], idx=0, memo={}):
    if idx > len(l) - 1:
        if debbug is True:
            print(memo)
        return h_pyear[0]
        
    if l[idx] - int(l[idx]) == 0:
        n_people += 1
    else:
        n_people -= 1
    
    if n_people >= h_pyear[1]:
        h_pyear[0], h_pyear[1] = l[idx], n_people
    
    memo[str(l[idx])] = n_people
    return year_most_people_alive(l, n_people, h_pyear, idx=idx + 1, memo=memo)



def main():
    max_age_ever = 70

    years = [[],[]]
    for n in range(100):
        byear = random.randint(1900, 2000)
        years[0].append(byear)


    for by in years[0]:
        dyear = random.randint(by, by + max_age_ever)
        years[1].append(dyear)

    # years = [[2000, 1975, 1975, 1803, 1750, 1840, 1803, 1894], [2010, 2005, 2003, 1809, 1869, 1935, 1921, 1921]]

    # years = [[2000, 2001],[2001, 2002]]
    if debbug is True:
        print_years(years)
        print_ages(years)
    nl = merge(years)
    r = year_most_people_alive(nl)
    print(r)


main()


```
